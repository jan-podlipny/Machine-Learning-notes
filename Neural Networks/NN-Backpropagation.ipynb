# Backpropagation Algorithm

$J(\Theta) = - \frac{1}{m} \sum_{i=1}^{m}\sum_{k=1}^{K}[y^{(i)}log(h_\theta(x^{(i)}_k) + (1 - y^{(i)}_k)log(1 - (h_\theta(x^{(i)})_k)] + \frac{\lambda}{2m} \sum_{l= 1}^{L-1} \sum_{i= 1}^{s_l} \sum_{j=1}^{s_{l}+1}
 (\theta_{j,i}^{(l)})^{2}$
 
 
**Goal**
- trying to minimize $\min\limits_{\Theta}$ $J(\Theta)$
**How** 
- need to compute:
    - $J(\Theta)$
    - $\frac{\partial}{\partial \Theta_{ij}^{l}}J(\Theta)$ (partial derivatives terms)


**Intuition**

Name comes from the fact, that we start computing delta terms for the output layer and then we go back a layer and compute the delta terms for the third hidden layer , and then we go back another step to delta 2 and so on



$\delta_{j}^{(l)}$ = "error" of node $j$ in layer $l$ 

For each output unit (layer L = 4)

$\delta_{j}^{(l)} = a_{j}^{(4)} - y_{j}$
- $a_{j}^{(4)} = (h_{\Theta}(x))$
    - delta term is difference between hypothesis output and what was the value of y in  training set, where $y_{j}$ is the *j* element of the vector y in  labeled training set
    
$\delta^{(3)} = (\Theta^{(3)})^T\delta^{4}. * g`(z^{(3)})$

$\delta^{(2)} = (\Theta^{(2)})^T\delta^{3}. * g`(z^{(2)})$
